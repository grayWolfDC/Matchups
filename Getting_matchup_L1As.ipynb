```python
import pandas as pd
```

```python
dfSwfVal = pd.read_csv('val1460128715760047_Rrs_mod.csv',header=0,index_col=0)
```

```python
def ConvertFileName(name):
    spl = name.split('L2')
    spl2 = spl[1].split('_OCVAL')
    return spl[0] + 'L1A' + spl2[0] + spl2[1]
```

```python
dfSwfVal['L1_name'] = dfSwfVal['file_name'].map(ConvertFileName)
```

```python
dfSwfVal.head()
```

```python
dfSwfVal.to_pickle('dfMatchUpSWF.pkl')
```

```python

import requests
for idx in dfSwfVal.index:
    fname = dfSwfVal.loc[idx,'L1_name']
    url = 'http://oceandata.sci.gsfc.nasa.gov/cgi/getfile/' + fname
    r = requests.get(url)
    with open(fname,'wb') as f:
        f.write(r.content)
```
