A look at 180+ SWF matchup data and comparison w/ corresponding MC uncertainty sim.

```python
import pickle
import pandas as pd
```

```python
import matplotlib.pyplot as pl
import seaborn as sb
```

```python
%matplotlib inline
```

```python
dfMupUncComp = pd.read_pickle('dfMupUncComp.pkl')
```

```python
dfMupUncComp.columns
```

```python
 dfUncs = dfMupUncComp.filter(regex='Rrs_unc',axis=1)
```

```python
xrng = (0,5e-3)
yrng = (0,xrng[1]/40)
dfUncs.plot(kind='scatter',x='Matchup_Rrs_unc_412',y='Rrs_unc_412',xlim=xrng,
           ylim=yrng)
```

```python
dfUncsStand=(dfUncs - dfUncs.mean()) / dfUncs.std()
```

```python
bands = ['412','443','490','510','555','670']
f,axs = pl.subplots(nrows=3,ncols=2,figsize=(10,12))
for band,ax in zip(bands,axs.ravel()):
    dfUncsStand.plot(kind='scatter',x='Matchup_Rrs_unc_'+band,y='Rrs_unc_'+band,
                    ax=ax)
```

```python
bands = ['412','443','490','510','555','670']
f,axs = pl.subplots(nrows=3,ncols=2,figsize=(10,12))
for band,ax in zip(bands,axs.ravel()):
    sb.regplot(x='Matchup_Rrs_unc_' + band, y='Rrs_unc_' + band,data=dfUncsStand,
               ax=ax)
    
```

```python
bands = ['412','443','490','510','555','670']
def JPlot(band):
    '''Plot STANDARDIZED uncertainty from Matchup vs Sensor noise for a given band '''
    sb.jointplot(x='Matchup_Rrs_unc_' + band, y='Rrs_unc_' + band,data=dfUncsStand,
                 kind='reg')
```
